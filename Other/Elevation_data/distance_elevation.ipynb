```python
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv("Sample_data.txt", sep="	")  
print(df.head())
x = df.latitude.values
y = df.longitude.values

distance = [0]
elevation = [0]
for i in range(1, x.shape[0]):
    delta_x = x[i] - x[i-1]
    delta_y = y[i] - y[i-1]
    rel_dist = (delta_x**2 + delta_y**2)**0.5
    elevation.append(rel_dist)
    distance.append(rel_dist + distance[i-1] )

df["elevation"]=elevation 
df["distance"]=distance

print()
print(df.head())

plt.plot(df["distance"], df["elevation"])
```
